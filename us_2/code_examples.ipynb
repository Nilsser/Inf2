**In python ist alles ein Pointer:** kopieren einer Liste

```python
fruits = ['apple', 'banana', 'cherry', 'orange']

x = fruits.copy()
fruits.append('grape')
print('x bleibt unverändert:', x,)
print('\nfruits enthält jetzt grape:', fruits)
```

**Listen "Wie ein Ring behandeln"**

```python
l= [1, 2, 3, 4, 5]
#nutze den modulo operator um von einem index n aus über die liste zu iterieren
n = 2
for i in range(2,2+len(l)):

    print(l[i % len(l)])
```

**Dictionary essentials**, für Crops&Dictionaries aufgabe

```python
# beispiel dictionary:
person = {'name': 'Alice', 'age': 30, 'city': 'New York'} #es gilt key:value

#iterieren über dictionary:
for key in person:
    print(key, ':', person[key])

#oder:
for key, value in person.items():
    print(key, ':', value,)

#erstellen eines neuen dictionarys (ohne comprehension):
squared_numbers = {}
for key in range(1, 6): #key ist hier die Zahl von 1 bis 5, welche wir quadrieren wollen.
    squared_numbers[key] = key ** 2
print(squared_numbers)
```

**Exceptions**

```python
try:
    eingabe = input("Bitte gib dein Alter ein: ")
    alter = int(eingabe)  # Hier passiert der Fehler, wenn es keine Zahl ist
    print(f"Cool, du bist ungefähr {alter*365} Tage alt.")

except ValueError:
    # Dieser Block wird nur ausgeführt, wenn int() scheitert
    print("Fehler: Das war keine Zahl! Bitte benutze nur Ziffern (0-9).")
```
